```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the CSV
df = pd.read_csv("Census_variables.csv")

# Clean up variable names
df["Variable"] = df["Variable"].replace({
    "White Ethnicity": "White",
    "No Qualification": "No Qual",
    "Level 4+ Qualification": "Level 4+",
    "House Ownership": "Own House",
    "Social Rent": "Social Rent",
    "No Car Access": "No Car"
})

# colorblind friendly
cud_colors = [
    "#0072B2",
    "#D55E00",
    "#009E73",
    "#E69F00",
    "#CC79A7",
    "#F0E442",
    "#56B4E9"
]

# Set plot style
sns.set(style="whitegrid", font_scale=1.1)

# Create the grouped bar chart
plt.figure(figsize=(12, 6))
ax = sns.barplot(data=df, x="Variable", y="Mean change", hue="Group", palette=cud_colors)

# Rotate x-axis labels for readability
plt.xticks(rotation=45, ha='right')

# Add axis labels and title
plt.ylabel("Mean % Change")
plt.xlabel("Socio-Economic Variable")
plt.title("Mean Change in Socio-Economic Variables Across LSOA Groups")

# Move legend outside the plot
plt.legend(title="Group", bbox_to_anchor=(1.05, 1), loc='upper left')

# Add horizontal grid lines
plt.grid(axis='y', linestyle='--', linewidth=0.5, alpha=0.7)

# Tight layout
plt.tight_layout()

# Show the plot
plt.show()


```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load and clean
df = pd.read_csv("Census_variables.csv")

# Rename variable labels for clarity
df["Variable"] = df["Variable"].replace({
    "White Ethnicity": "White",
    "No Qualification": "No qualifications",
    "Level 4+ Qualification": "Level 4+ qualifications",
    "House Ownership": "Owns home",
    "Social Rent": "Social Rent",
    "No Car Access": "No car access"
})

# Rename group labels for legend
df["Group"] = df["Group"].replace({
    "Leeds_2001_2021": "Leeds 20012021",
    "Top10_2001_2021": "Top 10 LSOAs 20012021"
})

# Filter for relevant groups
subset = df[df["Group"].isin(["Leeds 20012021", "Top 10 LSOAs 20012021"])]

# colorblind friendly
cb_palette = {
    "Leeds 20012021": "#0072B2",
    "Top 10 LSOAs 20012021": "#009E73"
}

# Plot
plt.figure(figsize=(10, 5))
sns.barplot(
    data=subset,
    x="Variable",
    y="Mean change",
    hue="Group",
    palette=cb_palette,
    dodge=True,
    width=0.5
)

plt.xticks(rotation=45, ha='right')
plt.ylabel("Mean % Change")
plt.xlabel("Socio-Economic Variable")
plt.title("Mean Change in Census Variables: Leeds vs Top 10 LSOAs (20012021)")
plt.legend(title="Group", bbox_to_anchor=(1.05, 1), loc='upper left')

# Add horizontal grid lines
plt.grid(axis='y', linestyle='--', linewidth=0.5, alpha=0.7)

plt.tight_layout()
plt.show()

```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load and clean
df = pd.read_csv("Census_variables.csv")

# Clean up variable names
df["Variable"] = df["Variable"].replace({
    "White Ethnicity": "White",
    "No Qualification": "No Qual",
    "Level 4+ Qualification": "Level 4+",
    "House Ownership": "Own House",
    "Social Rent": "Social Rent",
    "No Car Access": "No Car"
})

# Rename group labels for legend
df["Group"] = df["Group"].replace({
    "Leeds_2011_2021": "Leeds 20112021",
    "ELE_2011_2021": "ELE 20112021"
})

# Filter for ELE and Leeds
subset = df[df["Group"].isin(["Leeds 20112021", "ELE 20112021"])]

#  colorblind-friendly palette
cb_palette = {
    "ELE 20112021": "#D55E00",
    "Leeds 20112021": "#0072B2"
}

# Plot
plt.figure(figsize=(10, 5))
sns.barplot(
    data=subset,
    x="Variable",
    y="Mean change",
    hue="Group",
    palette=cb_palette,
    dodge=True,
    width=0.5
)

plt.xticks(rotation=45, ha='right')
plt.ylabel("Mean % Change")
plt.xlabel("Socio-Economic Variable")
plt.title("Mean Change in Census Variables: ELE LSOAs vs Leeds (20112021)")
plt.legend(title="Group", bbox_to_anchor=(1.05, 1), loc='upper left')

# Add horizontal grid lines
plt.grid(axis='y', linestyle='--', linewidth=0.5, alpha=0.7)

plt.tight_layout()
plt.show()

```
